Instantiate a `ColumnTransformer` named `transformer`, including both the `OneHotEncoder` and `OrdinalEncoder`. Be sure to specify the correct column names for each transformer.

<div class="alert alert-success"><u>Explanation for Exercice 5</u> 

**Step 1: Import the necessary library** 

This line imports the ColumnTransformer from sklearn.compose. A ColumnTransformer is used to apply different transformations to different subsets of columns in a DataFrame, allowing you to handle categorical, ordinal, and numerical data in parallel.

**Step 2: Define the categorical and ordinal columns**

We define two lists, categorical_columns and ordinal_columns, that specify which columns in X_imputed should be treated as categorical and which should be treated as ordinal.

**Step 3: Instantiate the ColumnTransformer** 

The ColumnTransformer is instantiated with two transformations:
- Categorical columns are transformed using the one_hot_encoder (previously defined). The one_hot_encoder will one-hot encode the columns in categorical_columns.
- Ordinal columns are transformed using the ordinal_encoder (previously defined). The ordinal_encoder will encode the columns in ordinal_columns with the specified order of categories.
The argument remainder='passthrough' ensures that any columns not explicitly mentioned (e.g., numerical columns or other features) are left unchanged and passed through to the output without transformation.

**Step 4: Fit and transform the X_imputed DataFrame** 

The fit_transform method is applied to the X_imputed DataFrame. This step performs the following:
- Fit: The one_hot_encoder learns the unique categories in the categorical columns, and the ordinal_encoder assigns ordinal rankings to the ordinal columns.
- Transform: The categorical columns are one-hot encoded, the ordinal columns are ordinal encoded, and any other columns (if they exist) are passed through unchanged.
The result of this step is a transformed matrix, where the categorical and ordinal columns have been encoded as numerical values, while other columns remain the same.
</div>

````python
# Step 1:
from sklearn.compose import ColumnTransformer

# Step 2:
categorical_columns=['Gender','Eye color','Race','Hair color','Skin color']
ordinal_columns=['Power Level','Intelligence Level']

# Step 3:
transformer = ColumnTransformer(
    transformers=
        [('categorical', one_hot_encoder, categorical_columns),
        ('ordinal', ordinal_encoder , ordinal_columns)],
    remainder='passthrough')

# Step 4:
X_encoded= transformer.fit_transform(X_imputed)
X_encoded 
````

- The team used a ColumnTransformer to apply both one-hot encoding to categorical columns and ordinal encoding to ordinal columns.

The pipeline processes categorical data:
- Imputation: Missing values are replaced using the most frequent category.
- One-hot Encoding: Categorical variables are transformed into binary columns, so that each unique category becomes its own feature.

</div>

````python
# Step 1: 
categorical_columns = ['Gender', 'Eye color', 'Race', 'Hair color']
numerical_columns = ['Height', 'Weight']
ordinal_columns = ['Power Level', 'Intelligence Level']
ordinal_categories = [
    ['Weak', 'Below Average', 'Average', 'Above Average', 'Extremely Powerful'],
    ['Low Intelligence', 'Average Intelligence', 'Smart', 'Genius', 'Super-Genius']
]

# Step 2:
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Step 3:
numerical_imputer = SimpleImputer(strategy='median')  # Replace missing numerical values with the median
categorical_imputer = SimpleImputer(strategy='most_frequent')  # Replace missing categorical values with the most frequent value
one_hot_encoder = OneHotEncoder(sparse_output=False)  
ordinal_encoder = OrdinalEncoder(categories=ordinal_categories)

# Step 4:
from sklearn.preprocessing import StandardScaler
numerical_pipeline = Pipeline(steps=[
    ('imputer', numerical_imputer), 
    ('scaler', StandardScaler()) 
]
)

# Step 5:
categorical_pipeline = Pipeline(steps=[
    ('imputer', categorical_imputer),
    ('onehot', one_hot_encoder)
]
)
````

<div class="alert alert-success"><u>Adding all the steps together </u> 

**Step 1: Importing ColumnTransformer**

We import ColumnTransformer from sklearn.compose, which allows us to apply different transformations to different subsets of features in parallel.

**Step 2: Defining the Preprocessor** 

The preprocessor combines the individual pipelines (numerical, categorical, and ordinal) and applies them to their respective columns in one unified step.
- 'num': Refers to the numerical pipeline that processes numerical_columns by imputing and scaling.
- 'cat': Refers to the categorical pipeline that processes categorical_columns by imputing and one-hot encoding.
- 'ord': Refers to the ordinal encoder that handles ordinal_columns based on predefined ordinal categories.
- remainder='passthrough': This option allows any columns not explicitly mentioned in the transformer to pass through unaltered.

</div>

````python
# Step 1:
from sklearn.compose import ColumnTransformer

# Step 2:
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_columns),
        ('cat', categorical_pipeline, categorical_columns),
        ('ord', ordinal_encoder, ordinal_columns)  # Ordinal columns processed separately
    ],
    remainder='passthrough'
)

preprocessor
````

<div class="alert alert-success"><u>Building the pipeline </u> 

**Step 1: Defining the Pipeline:**

- We create a pipeline object called pipe, which integrates all the preprocessing steps.
- The preprocessor step refers to a previously defined object that handles multiple preprocessing tasks like encoding, imputing missing data, and scaling. It could be a ColumnTransformer or a combination of encoders and transformers.

**Step 2: Fitting and Transforming the Dataset** 

The pipeline's fit_transform method is applied to the original feature matrix X. This step executes all the preprocessing steps defined in the pipeline

**Step 3: Ensuring the Correct Data Type** 

After applying the preprocessing steps, we convert the transformed data into a float data type. This is important because many machine learning models expect numerical inputs in floating-point format.
</div>

````python
# Step 1:
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor)  # Use the preprocessor defined above
])

# Step 2:
X_pipe = pipe.fit_transform(X)

# Step 3:
X_pipe=X_pipe.astype(float)
X_pipe
````
